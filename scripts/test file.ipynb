```python
from qmatch.extractors import get_extractor
from qmatch.data.classification import load_classification_datasets
```

```python
datasets_cls = load_classification_datasets()
X, y = datasets_cls["Iris_01"]
 
ext = get_extractor("classification")
result = ext.extract(X, y)
 
print(f"\n[Classification] Iris_01")
print(f"  vector shape : {result.vector.shape}")     # (24,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  task_type    : {result.task_type}")
print(f"  dim          : {result.dim}")
print(f"  to_dict()    : { {k: round(v, 4) for k, v in list(result.to_dict().items())[:3]} }...")
```

```python
from qmatch.data.regression import load_regression_datasets
 
datasets_reg = load_regression_datasets()
name_reg = list(datasets_reg.keys())[0]
X, y = datasets_reg[name_reg]
 
ext = get_extractor("regression")
result = ext.extract(X, y)
 
print(f"\n[Regression] {name_reg}")
print(f"  vector shape : {result.vector.shape}")     # (19,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  dim          : {result.dim}")
```

```python
from qmatch.data.timeseries import load_timeseries_datasets
 
datasets_ts = load_timeseries_datasets()
name_ts = list(datasets_ts.keys())[0]
X, y = datasets_ts[name_ts]
 
ext = get_extractor("timeseries")
result = ext.extract(X, y)
 
print(f"\n[TimeSeries] {name_ts}")
print(f"  X shape      : {X.shape}")                 # (n_samples, n_timesteps)
print(f"  vector shape : {result.vector.shape}")     # (25,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  dim          : {result.dim}")
```

```python
from qmatch.data.multilabel import load_multilabel_datasets
 
datasets_ml = load_multilabel_datasets()
name_ml = list(datasets_ml.keys())[0]
X, Y = datasets_ml[name_ml]
 
ext = get_extractor("multilabel")
result = ext.extract(X, Y)            # Y là 2D matrix, truyền vào vị trí y
 
print(f"\n[Multi-label] {name_ml}")
print(f"  X shape      : {X.shape}")
print(f"  Y shape      : {Y.shape}")                 # (n_samples, n_labels)
print(f"  vector shape : {result.vector.shape}")     # (15,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  dim          : {result.dim}")
```

```python
from qmatch.data.clustering import load_clustering_datasets
 
datasets_clu = load_clustering_datasets()
name_clu = list(datasets_clu.keys())[0]
X = datasets_clu[name_clu]            # không có y
 
ext = get_extractor("clustering")
result = ext.extract(X)               # không truyền y
 
print(f"\n[Clustering] {name_clu}")
print(f"  X shape      : {X.shape}")
print(f"  vector shape : {result.vector.shape}")     # (14,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  dim          : {result.dim}")
```

```python
from qmatch.data.anomaly import load_anomaly_detection_datasets
 
datasets_ad = load_anomaly_detection_datasets()
name_ad = list(datasets_ad.keys())[0]
X, y = datasets_ad[name_ad]
 
ext = get_extractor("anomaly")
result = ext.extract(X, y)
 
print(f"\n[Anomaly] {name_ad}")
print(f"  X shape      : {X.shape}")
print(f"  contamination: {(y == 1).mean():.4f}")
print(f"  vector shape : {result.vector.shape}")     # (17,)
print(f"  feature_names: {result.feature_names[:5]}...")
print(f"  dim          : {result.dim}")
```
